```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

### Definição de variáveis (simulando o efeito de um martingale)

```python
# Capital Inicial
capital = 100

# Quanto que será apostado em cada iteração
bet_size = 10

# Chance de ganhar
prob_win = 0.5

# Change de perder
prob_loos = (1 - prob_win)

# Quanto será ganho em caso de vitória
win_reward = 1

# Quanto será perdido em caso de derrota
loss_reward = -1
```

### Execução do algoritmo

```python
fig, ax = plt.subplots(figsize=(10, 5))

# Curva de equidade
equity_curve = [capital]

# Resultado final da simulação
last_result = []
for j in range(100):
    # Reseta a curva de equidade
    bet_size = 10
    equity_curve = [capital]
    for i in range(1000):
        # Se a probabilidade de perder for menor que o número aleatório (entre 0 e 1), então o último resultado será o valor da curva de equidade mais o valor da aposta
        if prob_loos < random.random():
            equity_curve += [equity_curve[-1] + win_reward * bet_size]
            # Se ganhar a aposta, o valor da aposta será zerado
            bet_size = 10
        else:
            equity_curve += [equity_curve[-1] + loss_reward * bet_size]
            # Se perder a aposta, o valor da aposta será dobrado
            bet_size = bet_size * 2
        if equity_curve[-1] <= 0:
            break
    
    # Adiciona o resultado final da simulação (valor recebido ou perdido) na lista de resultados finais    
    last_result += [equity_curve[-1] - capital]

    ax.plot(equity_curve)
```

### Resultado Final
Valor obtido ao executar 1000 cenários

```python
# Média dos resultados finais
np.mean(last_result)
```

É possível verificar que com o margin gale, o valor obtido ao final das simulações pode ser bem diferente de 0. Acontece que, muitas situações levam a uma quantidade de martin gales que pode ultrapassar o valor de capital, levando a ruína
